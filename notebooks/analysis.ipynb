```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

df = pd.read_csv("../data/raw/telco_customer_churn_dataset.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.countplot(x="Churn", data=df)
plt.title("Churn Distribution")
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 4))

sns.histplot(df["MonthlyCharges"], bins=30, ax=ax[0])
ax[0].set_title("Monthly Charges (Training Data)")

sns.histplot(df["tenure"], bins=30, ax=ax[1])
ax[1].set_title("Tenure (Training Data)")

plt.tight_layout()
plt.show()
```

### Monitoring Motivation

In real-world production systems, customer behavior changes over time.
Key numerical features such as tenure and monthly charges are prone to
distribution shifts, which can degrade model performance if not monitored.

This project explicitly monitors such shifts to trigger retraining when necessary.
